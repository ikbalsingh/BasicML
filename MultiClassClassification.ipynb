```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import preprocessing,linear_model,svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
%matplotlib inline
```

```python
X = pd.read_csv('../Desktop/machine-learning-ex3/ex3/X.csv', sep=",", header=None)
y = pd.read_csv('../Desktop/machine-learning-ex3/ex3/Y.csv', sep=",", header=None)
```

```python
sample = np.array(X.sample(frac=0.5))
# sample = preprocessing.scale(sample)
sample.shape
for i in range(100):
    intensitymatrix = np.array(sample[i]).reshape(20,20)
    ax = plt.subplot(10,10,i+1)
    im = ax.imshow(intensitymatrix, cmap='gray')
    ax.set_xticks([])
    ax.set_yticks([])
plt.show()
```

```python
X = np.array(X)
y = np.array(y)
# X = preprocessing.scale(X)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)
```

```python
clf = linear_model.LogisticRegression()
# clf = svm.SVC()
clf.fit(X_train,y_train.ravel())
```

```python
pred = clf.predict(X_test)
accuracy_score(y_test,pred)
```

```python
from sklearn.neural_network import MLPClassifier
clfa = MLPClassifier()

clfa.fit(X_train,y_train.ravel())
```

```python
pred = clfa.score(X_test,y_test)
pred
```
